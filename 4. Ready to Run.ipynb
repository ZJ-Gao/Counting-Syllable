```python
from requests_html import HTMLSession
session = HTMLSession()
```

```python
# Read document and split them into a list
# opening the text file 
fp = open('Middle.txt')
for word in fp:
    content = word
fp.close()
list = content.split()
list
```

```python
for word in list:
    url = 'https://www.writerlywords.com/syllables/'+word
    print(url)
    # Catch information from website
    r = session.get(url)
    sel = '#ctl00_ContentPane_paragraphtext'
    results = r.html.find(sel)
    text = str(results[0].text)
    # write text into txt document
    text_file = open("ResultMiddle.txt", "a")
    n = text_file.write('\n')
    n = text_file.write(text)
    text_file.close()
```
